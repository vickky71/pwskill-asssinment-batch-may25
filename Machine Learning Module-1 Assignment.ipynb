## Question 1: Explain the differences between AI, ML, Deep Learning (DL), and Data Science (DS).

## Question 2: What are the types of machine learning? Describe each with one real-world example.

### Answer: 1. Supervised Learning

**Definition:** The algorithm learns from **labeled data**—each input is paired with a correct output—and uses that to predict outcomes for new, unseen data. 
**Example:** Detecting spam emails. A model trained with emails clearly labeled "spam" or "not spam" learns to classify new incoming emails based on features like sender, content, and metadata.

### 2. Unsupervised Learning

**Definition:** The algorithm works with **unlabeled data**, identifying patterns, structures, or groupings on its own. 
**Example:** Customer segmentation in retail. Without pre-set labels, the algorithm groups customers based on buying behavior, helping tailor marketing strategies.

### 3. Reinforcement Learning

**Definition:** An **agent interacts with an environment** and learns to make decisions through **rewards or penalties**. 
**Example:** Self-driving cars. Systems continuously observe the driving environment and learn maneuvers—like stopping or turning—based on rewards for safe driving behaviors. 

### 4. Semi‑Supervised Learning

**Definition:** A middle ground where the model is trained on **a small amount of labeled data** along with **a larger set of unlabeled data**. 
**Example:** Improving language understanding in search. Google uses semi‑supervised methods: limited labeled queries guide the model, while the larger pool of unlabeled text helps enhance overall comprehension. 


### 5. Self‑Supervised Learning

**Definition:** This method lets models **generate their own supervision** from unlabeled data by predicting parts of the inputs from other parts—bridging unsupervised and supervised learning.
**Example:** Training large language models (like GPT‑3/GPT‑4). They learn to predict the next word in a sentence using massive amounts of unlabeled text, effectively teaching themselves nuances of language. 


### 6. Ensemble Learning (Bonus)

**Definition:** Not a standalone learning paradigm, but a technique that **combines multiple models**—like bagging or boosting—to improve accuracy and robustness.
**Example:** Fraud detection systems. By combining several decision trees or classifiers (as in random forests or boosting), the system becomes more reliable at identifying suspicious, fraudulent transactions.


**Summary Table:**

| Type                   | What It Uses               | Real‑World Example                     |
| ---------------------- | -------------------------- | -------------------------------------- |
| Supervised Learning    | Labeled data               | Spam detection                         |
| Unsupervised Learning  | Unlabeled data             | Customer segmentation                  |
| Reinforcement Learning | Environment + feedback     | Self‑driving cars                      |
| Semi‑Supervised        | Mixed labeled + unlabeled  | Search query understanding             |
| Self‑Supervised        | Predicting parts of inputs | Language model pretraining (e.g., GPT) |
| Ensemble Learning      | Multiple models            | Fraud detection with aggregated models

## Question 3: Define overfitting, underfitting, and the bias-variance tradeoff in machine learning.

## Answer: Overfitting

**Definition:**
Overfitting occurs when a model fits the **training data too closely**, capturing not only the true underlying patterns but also the noise and outliers. Such a model performs exceptionally well on training data but poorly on new, unseen data.

**Key Characteristics:**

* **Low bias, high variance** — the model is too flexible and overly sensitive to fluctuations in training data. 
* Often happens when the model is overly complex (e.g., high-degree polynomials, deep neural networks without regularization).

## Underfitting

**Definition:**
Underfitting happens when the model is **too simplistic** to capture the underlying structure in the data. It fails to learn important patterns and performs poorly on both training and test sets.

**Key Characteristics:**

* **High bias, low variance** — the model makes strong assumptions and lacks flexibility.
* Examples include using a linear model to fit data that exhibits a nonlinear trend.

## Bias–Variance Tradeoff

**Definition:**
This tradeoff refers to the balance between two types of errors that affect model generalization:

* **Bias**: Error due to assumptions in the learning algorithm, leading to underfitting.
* **Variance**: Error due to sensitivity to fluctuations in training data, leading to overfitting.

The total expected error on unseen data can be decomposed as:

$$
\text{Error} = \text{Bias}^2 + \text{Variance} + \text{Irreducible Error}
$$

The goal is to find a sweet spot where both bias and variance are minimized, enabling the model to generalize well. 

**Visual Analogy:**

* A **simple model** yields **high bias**, **low variance** → underfits.
* A **complex model** yields **low bias**, **high variance** → overfits.
* An **optimal model** balances both, achieving good generalization.

| Overfitting  | Model too complex, captures noise              | Low bias, high variance | Excellent train accuracy; poor on new data       |

### Additional Insights

* **Overfitting**: A model "memorizes" training data instead of learning general trends. It might achieve near-perfect training accuracy but generalize poorly. 
* **Underfitting**: Reflects an inability to capture complexity even in training data—often due to an overly simplistic model.
* **Bias–Variance Tradeoff in Practice**: Adjusting model complexity (e.g., changing polynomial degree), using regularization, cross-validation, or ensemble techniques helps navigate this tradeoff.

## Question 4: What are outliers in a dataset, and list three common techniques for handling them?

### Answer: Outliers are data points that stand out significantly from the rest—think of a monthly sales figure that's dramatically higher or lower than normal. They can arise from errors (like a typo or sensor glitch) or represent genuine, rare events. While sometimes harmless or even meaningful, outliers can often skew analyses or mislead machine learning models.

### Three Solid Techniques to Handle Outliers

1. **IQR Method (Interquartile Range)**

   * **How it works:** Calculate the lower (Q1) and upper (Q3) quartiles, then flag anything below Q1 − 1.5×IQR or above Q3 + 1.5×IQR.
   * **Why use it:** It’s simple, distribution-free, and works well for skewed or unknown data distributions.

2. **Z‑Score Method**

   * **How it works:** Compute how many standard deviations each point is from the mean. Typically, points with |z| > 3 are considered outliers.
   * **Why use it:** It’s intuitive and effective for datasets that follow (or approximate) a normal distribution.

3. **Winsorizing (Capping Extreme Values)**

   * **How it works:** Instead of dropping or ignoring outliers, you bring them inward by replacing extreme values with the nearest cutoff—like setting anything above the 95th percentile to the 95th percentile.
   * **Why use it:** You maintain dataset size while mitigating the distortion caused by extreme values.


### Quick Comparison

| Method      | Approach                            | Best For                                 |
| ----------- | ----------------------------------- | ---------------------------------------- |
| IQR Rule    | Beyond Q1 – 1.5×IQR or Q3 + 1.5×IQR | Skewed or unknown distributions          |
| Z‑Score     | Outside ±3 standard deviations      | Normally distributed data                |
| Winsorizing | Cap extremes to threshold           | Keeping all data but reducing distortion |


### Why it Matters

Consider a few fiery temperature readings due to equipment error—these are outliers that can break your model. Maybe they should be corrected or removed. On the other hand, if you're analyzing income data and spot a few very high earners, those might be pure gold — especially for detecting fraud. The treatment choice matters.

## Question 5: Explain the process of handling missing values and mention one imputation technique for numerical and one for categorical data.

### Answer:Process Overview: Handling Missing Values

1. **Understand the Missingness**
   Begin by exploring your dataset to determine why values are missing—whether due to measurement errors, non-responses, or systemic issues. Knowing if data is *Missing Completely At Random (MCAR)*, *Missing At Random (MAR)*, or *Missing Not At Random (MNAR)* guides your strategy.
   
3. **Choose a Strategy**

   * **Deletion**: If missing values are rare and random, you might drop those records—but use sparingly to avoid losing important information.
   * **Imputation**: Typically preferable—replace missing entries using statistical or model-based methods to retain dataset size and integrity.

4. **Impute with Appropriate Methods**
   Select methods based on data type and distribution. Simple techniques are quick and effective; advanced methods capture more signal but require more effort.


### Imputation Techniques

* **Numerical Data – Median Imputation**
  Replace missing numerical values with the **median**, the middle value of observed data. This method minimizes the impact of outliers and is robust for skewed distributions.

* **Categorical Data – Mode Imputation**
  Fill missing categorical values with the **mode**—the most frequent category. It’s straightforward and preserves the natural distribution of categories.

### Summary Table

| Step                 | Action & Why                                                              |
| -------------------- | ------------------------------------------------------------------------- |
| Explore Missingness  | Identify pattern/type of missing data (MCAR, MAR, MNAR)                   |
| Choose Strategy      | Prefer imputation over deletion when possible to keep data integrity      |
| Impute (Numerical)   | Use **median imputation** for robustness against outliers and skewed data |
| Impute (Categorical) | Use **mode imputation** to maintain categorical consistency               |


```python
import pandas as pd
import numpy as np

# Create a sample DataFrame with missing values (NaN)
df = pd.DataFrame({
    'Age': [25, 30, np.nan, 22, 28],
    'Salary': [50000, 60000, 55000, np.nan, 58000],
    'City': ['Delhi', None, 'Mumbai', 'Delhi', None]
})

# Check the DataFrame before imputation
print("Before imputation:")
print(df)

# Median imputation for numeric columns
df['Age'] = df['Age'].fillna(df['Age'].median())
df['Salary'] = df['Salary'].fillna(df['Salary'].median())

# Mode imputation for categorical column (using mode()[0])
df['City'] = df['City'].fillna(df['City'].mode()[0])

# Check the DataFrame after imputation
print("\nAfter imputation:")
print(df)
```

## Question 8: Write a Python script to:

● Generate 1000 samples from a normal distribution.

● Introduce 50 random missing values.

● Fill missing values with the column mean.

● Plot a histogram before and after imputation.

(Include your Python code and output in the code box below.)


```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# Step 1: Generate 1000 samples from a normal distribution
np.random.seed(42)  # For reproducibility
data = np.random.normal(loc=0, scale=1, size=1000)
df = pd.DataFrame({'Value': data})

# Step 2: Introduce 50 random missing values
missing_indices = np.random.choice(df.index, size=50, replace=False)
df.loc[missing_indices, 'Value'] = np.nan

# Step 3: Plot histogram before imputation
plt.figure(figsize=(12, 6))

plt.subplot(1, 2, 1)
df['Value'].dropna().hist(bins=30, color='skyblue', edgecolor='black')
plt.title('Before Imputation')
plt.xlabel('Value')
plt.ylabel('Frequency')

# Step 4: Fill missing values with the column mean
mean_value = df['Value'].mean()
df_imputed = df.copy()
df_imputed['Value'].fillna(mean_value, inplace=True)

# Step 5: Plot histogram after imputation
plt.subplot(1, 2, 2)
df_imputed['Value'].hist(bins=30, color='lightgreen', edgecolor='black')
plt.title('After Imputation (Mean)')
plt.xlabel('Value')
plt.ylabel('Frequency')

plt.tight_layout()
plt.show()
```

## Question 10: You are working as a data scientist for a retail company. You receive a customer transaction dataset that contains:

● Missing ages,

● Outliers in transaction amount,

● A highly imbalanced target (fraud vs. non-fraud),

● Categorical variables like payment method.

Explain the step-by-step data preparation plan you’d follow before training a machine learning
model. Include how you’d address missing data, outliers, imbalance, and encoding.
(Include your Python code and output in the code box below.)


## Answer:Step-by-Step Data Preparation Plan 
**1.Handle Missing Data**

* Numerical Features: Impute missing values with the mean.

* Categorical Features: Impute missing values with the most frequent category.

**2.Detect and Handle Outliers**

* Use the Interquartile Range (IQR) method to identify and cap outliers in the transaction amount.

**3.Address Class Imbalance**

* Apply SMOTE (Synthetic Minority Over-sampling Technique) to oversample the minority class (fraud).

**4.Encode Categorical Variables**

* Use One-Hot Encoding for categorical variables like payment method.


```python
import pandas as pd, numpy as np
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import OneHotEncoder
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import classification_report

# Try importing SMOTE; if unavailable, proceed without it
try:
    from imblearn.over_sampling import SMOTE
    smote = SMOTE(random_state=42)
    smote_status = "SMOTE applied"
except ImportError:
    smote = None
    smote_status = "SMOTE unavailable"

# Synthetic retail fraud dataset
df = pd.DataFrame({
    'Age': [25, np.nan, 30, 22, np.nan, 35, 40, 28, np.nan, 50],
    'TransactionAmount': [200, 5000, 300, 150, 100000, 400, 250, 600, 1500, 800],
    'PaymentMethod': ['Credit', 'Debit', 'Credit', 'Cash', 'Debit', 'Credit', 'Cash', 'Credit', 'Debit', 'Cash'],
    'Fraud': [0, 1, 0, 0, 1, 0, 0, 0, 1, 0]
})

# 1. Impute missing ages using median
df['Age'] = SimpleImputer(strategy='median').fit_transform(df[['Age']])

# 2. Cap outliers in TransactionAmount using IQR
Q1, Q3 = df['TransactionAmount'].quantile([0.25, 0.75])
IQR = Q3 - Q1
df['TransactionAmount'] = df['TransactionAmount'].clip(Q1 - 1.5*IQR, Q3 + 1.5*IQR)

# 3. One-hot encode 'PaymentMethod'
ohe = OneHotEncoder(sparse=False, drop='first')
pm = ohe.fit_transform(df[['PaymentMethod']])
df = pd.concat([df.drop('PaymentMethod', axis=1), pd.DataFrame(pm, columns=ohe.get_feature_names_out())], axis=1)

# Prepare features and target
X, y = df.drop('Fraud', axis=1), df['Fraud']

# 4. Apply SMOTE if available
if smote:
    X, y = smote.fit_resample(X, y)
print(smote_status, "| Class distribution:", np.bincount(y))

# 5. Train-test split, model training, evaluation
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42, stratify=y)
model = RandomForestClassifier(random_state=42).fit(X_train, y_train)
print("\nClassification Report:\n", classification_report(y_test, model.predict(X_test)))
```
