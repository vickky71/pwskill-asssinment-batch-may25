## Question 1: What is the difference between descriptive statistics and inferential statistics? Explain with examples.

### Answer-**Introduction:**

Statistics is a branch of mathematics that deals with collecting, analyzing, interpreting, and presenting data. It is mainly divided into two categories: **Descriptive Statistics** and **Inferential Statistics**. Both play a crucial role in data analysis but serve different purposes.

### **(I) Descriptive Statistics:**

**Definition:**
Descriptive statistics involves **summarizing and organizing** data so that it can be easily understood. It describes the **main features** of a dataset **without making any predictions or generalizations**.

**Key Features:**

* Deals with **actual data**
* No assumptions or predictions are made
* Provides measures such as:

  * Central tendency (mean, median, mode)
  * Dispersion (range, variance, standard deviation)
  * Data distribution (frequency, percentages, graphs)

**Example:**
If you conducted a survey on the heights of 100 students, descriptive statistics would help you find:

* Average height = 162.5 cm
* Tallest student = 180 cm
* Shortest student = 150 cm

These values describe the characteristics of the **sample only**.

### **(II) Inferential Statistics:**

**Definition:**
Inferential statistics involves **making predictions or inferences** about a larger population **based on a sample** of data. It uses **probability theory** to estimate and test hypotheses.

**Key Features:**

* Works with **samples** to draw conclusions about populations
* Includes:

  * Hypothesis testing
  * Confidence intervals
  * Regression analysis
  * ANOVA
* Helps in **decision-making** under uncertainty

**Example:**
Suppose you surveyed 100 voters out of a city of 10,000. If 60% of your sample supports a candidate, inferential statistics helps you estimate, with confidence, that **about 60% of the entire population may support the same candidate**, within a **margin of error**.

You might report:

> “We are 95% confident that between 55% and 65% of the voters in the city support the candidate.”


### **Key Differences:**

| Feature              | Descriptive Statistics               | Inferential Statistics                          |
| -------------------- | ------------------------------------ | ----------------------------------------------- |
| **Purpose**          | To describe and summarize data       | To infer or predict from sample to population   |
| **Basis**            | Uses entire data set                 | Uses sample data                                |
| **Example Outcome**  | Mean age of 100 students is 21 years | Estimate average age of all students in college |
| **Techniques**       | Mean, median, mode, charts           | Hypothesis tests, confidence intervals          |
| **Conclusion Scope** | Limited to the dataset used          | Generalized to a larger population              |


### **Conclusion:**

Descriptive statistics provides a **snapshot** of the data, while inferential statistics helps make **predictions and informed decisions** based on that snapshot. A good data analyst must know when and how to use both depending on the problem and data available.

### **Question 2: What is sampling in statistics? Explain the differences between random and stratified sampling.**

### Answer- **Introduction:**

In statistics, working with an entire population is often impractical due to time, cost, and effort. Instead, we select a **subset of the population**, called a **sample**, to analyze and draw conclusions. This process is known as **sampling**.

### **What is Sampling?**

#### **Definition:**

Sampling is the **process of selecting a portion (sample)** from a larger group (population) in such a way that the selected sample **represents** the population well.

#### **Purpose of Sampling:**

* To reduce cost and time
* To make data collection feasible
* To allow statistical inference about the population

#### **Key Terms:**

* **Population:** The entire group we are interested in studying (e.g., all college students in India)
* **Sample:** A smaller group selected from the population (e.g., 500 students from different states)
* **Sampling Method:** The way we choose the sample (e.g., random, stratified)

### **1. Random Sampling:**

#### **Definition:**

In **random sampling**, every member of the population has an **equal chance** of being selected. It is purely **by chance**, without any specific pattern or grouping.

#### **Features:**

* Simple and unbiased
* Each individual is selected independently
* Used when the population is **homogeneous**

#### **Example:**

Suppose a teacher wants to choose 10 students randomly from a class of 50 for a survey. They assign numbers to each student and use a random number generator to pick 10 students.

### **2. Stratified Sampling:**

#### **Definition:**

In **stratified sampling**, the population is first **divided into groups** (called **strata**) based on certain characteristics, and then a **random sample is taken from each group**.

#### **Features:**

* Ensures representation from all groups
* Reduces sampling error
* Useful when the population is **heterogeneous**

#### **Example:**

Suppose a college has 500 students: 300 undergraduates and 200 postgraduates. If we want a sample of 50 students using stratified sampling, we might randomly select:

* 30 students from undergraduates (60%)
* 20 students from postgraduates (40%)

This way, both groups are proportionally represented.

### **Key Differences Between Random and Stratified Sampling:**

| Feature              | Random Sampling                           | Stratified Sampling                                  |
| -------------------- | ----------------------------------------- | ---------------------------------------------------- |
| **Basis**            | Entire population is treated as one group | Population is divided into strata (subgroups)        |
| **Selection Method** | Random selection from full population     | Random selection **within each group**               |
| **Representation**   | May not represent all subgroups equally   | Ensures proportional representation of each group    |
| **Best For**         | Homogeneous population                    | Heterogeneous population                             |
| **Example**          | 10 students chosen randomly from class    | 5 boys and 5 girls chosen from class (gender strata) |


### **Conclusion:**

Sampling is a crucial concept in statistics used to study large populations efficiently. Both **random** and **stratified sampling** have their own advantages depending on the nature of the population. While random sampling is simpler and unbiased, **stratified sampling gives more accurate and reliable results** when the population has distinct groups. Choosing the right sampling method is key to ensuring valid statistical analysis.

### **Question 3: Define mean, median, and mode. Explain why these measures of central tendency are important.**

### Answer- **Introduction:**

In statistics, **measures of central tendency** are used to describe the center or average of a dataset. The three most common measures are **mean**, **median**, and **mode**. These values help summarize a large dataset with a single representative number and are essential for understanding the distribution of data.

### **1. Mean (Arithmetic Average):**

#### **Definition:**

The **mean** is the sum of all values in a dataset divided by the total number of values.

#### **Formula:**

$$
\text{Mean} = \frac{\text{Sum of all values}}{\text{Number of values}}
$$

#### **Example:**

If the marks of 5 students are: 70, 80, 90, 60, and 100
Mean = (70 + 80 + 90 + 60 + 100) ÷ 5 = 400 ÷ 5 = **80**

#### **Key Points:**

* Most commonly used measure
* Sensitive to extreme values (outliers)


### **2. Median (Middle Value):**

#### **Definition:**

The **median** is the middle value of a dataset when the values are arranged in **ascending or descending** order.

* If the number of observations is **odd**, median is the middle value.
* If the number is **even**, median is the average of the two middle values.

#### **Example:**

Data: 40, 50, 60, 70, 80 → Median = **60** (middle value)
Data: 30, 40, 50, 60 → Median = (40+50)/2 = **45**

#### **Key Points:**

* Not affected by outliers
* Best for skewed distributions

### **3. Mode (Most Frequent Value):**

#### **Definition:**

The **mode** is the value that occurs **most frequently** in the dataset.

#### **Example:**

Data: 10, 20, 20, 30, 40 → Mode = **20** (appears twice)
Data: 5, 5, 5, 6, 6, 7, 8 → Mode = **5**

#### **Key Points:**

* There can be more than one mode (bi-modal, multi-modal)
* Can be used for **categorical data** (e.g., most preferred brand)


### **Importance of Measures of Central Tendency:**

| Importance                     | Explanation                                                                      |
| ------------------------------ | -------------------------------------------------------------------------------- |
| **Simplifies data**            | Reduces a large dataset to one representative value                              |
| **Easy comparison**            | Allows comparison between different datasets or groups                           |
| **Foundation of analysis**     | Forms the basis of many statistical techniques like variance, standard deviation |
| **Decision making**            | Helps businesses, researchers, and policymakers to make informed decisions       |
| **Understanding distribution** | Gives insight into the nature and shape of data (e.g., normal, skewed)           |



### **When to Use Which?**

| Situation                           | Best Measure | Reason                             |
| ----------------------------------- | ------------ | ---------------------------------- |
| Data has extreme values (outliers)  | **Median**   | Not affected by outliers           |
| Data is evenly distributed          | **Mean**     | Utilizes all values                |
| Categorical or non-numeric data     | **Mode**     | Most suitable for non-numeric data |
| You want the most frequent response | **Mode**     | Highlights popularity or trends    |



### **Conclusion:**

Mean, median, and mode are the fundamental statistical tools used to **summarize data efficiently**. Each measure has its own advantages and is suitable for different types of data and analysis goals. Understanding and correctly applying these measures helps in making data-driven decisions in fields like education, economics, healthcare, and business.

### **Question 4: Explain skewness and kurtosis. What does a positive skew imply about the data?**

### Answer- **Introduction:**

In statistics, it’s important not just to know the average (mean) or the spread (standard deviation) of a dataset — but also the **shape of the distribution**. Two important concepts that describe this shape are:

* **Skewness**: Describes the **asymmetry** of the distribution.
* **Kurtosis**: Describes the **peakedness** or **flatness** of the distribution.

These help us understand the **behavior of data beyond central tendency and spread**.

### **1. Skewness:**

#### **Definition:**

Skewness measures the **degree of asymmetry** of a distribution around its mean.

* **Symmetrical distribution** has skewness = 0
* **Positive skew** means the right tail is longer (more extreme high values)
* **Negative skew** means the left tail is longer (more extreme low values)

#### **Types of Skewness:**

| Type              | Description                                                | Visual Shape      |
| ----------------- | ---------------------------------------------------------- | ----------------- |
| **Positive Skew** | Tail on the **right side** is longer; mean > median > mode | Long right tail   |
| **Negative Skew** | Tail on the **left side** is longer; mean < median < mode  | Long left tail    |
| **No Skew**       | Symmetrical; mean = median = mode                          | Bell-shaped curve |

#### **Example (Positive Skew):**

Income data often shows positive skew — a few people earn extremely high income, while most people earn average or below-average income.


### **2. Kurtosis:**

#### **Definition:**

Kurtosis measures the **"tailedness"** or **peakedness** of a distribution — that is, how sharp or flat the peak is and how heavy the tails are.

#### **Types of Kurtosis:**

| Type            | Description                                      | Visual Shape               |
| --------------- | ------------------------------------------------ | -------------------------- |
| **Mesokurtic**  | Normal distribution (benchmark), kurtosis ≈ 3    | Moderate peak and tails    |
| **Leptokurtic** | More peaked and **fatter tails**, kurtosis > 3   | Sharp peak, more outliers  |
| **Platykurtic** | Flatter peak and **thinner tails**, kurtosis < 3 | Broad peak, fewer outliers |

#### **Example:**

* Exam scores that are tightly clustered around the mean with a few extreme outliers → **leptokurtic**
* Uniform data with no extreme values → **platykurtic**


### **What Does a Positive Skew Imply About the Data?**

A **positive skew** (also called **right-skewed distribution**) implies:

* The **tail on the right side** is longer
* Most data values are **concentrated on the left**
* There are **a few extremely high values** that pull the mean upward
* The **mean > median > mode**

#### **Interpretation:**

A positive skew indicates that while **most observations are low or average**, a **few high values** are increasing the mean.

#### **Example:**

If you analyze the salaries in a company:

* Most employees may earn ₹30,000–₹50,000
* A few executives may earn ₹2,00,000+
* This causes a **positive skew**, increasing the average salary, though most earn below that average


### **Importance in Data Analysis:**

| Concept      | Why It's Important                                                                           |
| ------------ | -------------------------------------------------------------------------------------------- |
| **Skewness** | Helps identify **data asymmetry**; guides selection of statistical methods                   |
| **Kurtosis** | Helps detect **outliers** and **risk** in financial data; affects reliability of predictions |


### **Conclusion:**

Understanding **skewness and kurtosis** is essential in statistics to interpret the shape and nature of the data distribution. While skewness tells us about **asymmetry**, kurtosis tells us about **the sharpness and tail heaviness** of the distribution. A **positive skew** signals that the data has **few high outliers** and most values lie on the lower side — a vital insight for accurate data modeling and decision-making.

### Question 5: Implement a Python program to compute the mean, median, and mode of a given list of numbers.
### numbers = [12, 15, 12, 18, 19, 12, 20, 22, 19, 19, 24, 24, 24, 26, 28]
### (Include your Python code and output in the code box below.)

## **Answer**-


```python
# Python program to compute the mean, median, and mode
import statistics

# Given list of numbers
numbers = [12, 15, 12, 18, 19, 12, 20, 22, 19, 19, 24, 24, 24, 26, 28]

# Calculate Mean
mean = statistics.mean(numbers)

# Calculate Median
median = statistics.median(numbers)

# Calculate Mode
mode = statistics.mode(numbers)

# Display the results
print("Mean:", mean)
print("Median:", median)
print("Mode:", mode)
```

### **Question 6**- Compute the covariance and correlation coefficient between the following
### two datasets provided as lists in Python:
### list_x = [10, 20, 30, 40, 50]
### list_y = [15, 25, 35, 45, 60]
### (Include your Python code and output in the code box below.)

## **Answer-**

```python
# Python program to compute covariance and correlation coefficient
import numpy as np

# Given datasets
list_x = [10, 20, 30, 40, 50]
list_y = [15, 25, 35, 45, 60]

# Convert lists to NumPy arrays
x = np.array(list_x)
y = np.array(list_y)

# Compute Covariance Matrix
cov_matrix = np.cov(x, y, bias=False)

# Extract covariance value
covariance = cov_matrix[0][1]

# Compute Correlation Coefficient
correlation = np.corrcoef(x, y)[0][1]

# Display the results
print("Covariance:", covariance)
print("Correlation Coefficient:", correlation)
```

### Question 7: Write a Python script to draw a boxplot for the following numeric list and identify its outliers. Explain the result:
### data = [12, 14, 14, 15, 18, 19, 19, 21, 22, 22, 23, 23, 24, 26, 29, 35]
### (Include your Python code and output in the code box below.)

## **Answer-**


```python
# Python script to draw a boxplot and identify outliers
import matplotlib.pyplot as plt
import numpy as np

# Given data
data = [12, 14, 14, 15, 18, 19, 19, 21, 22, 22, 23, 23, 24, 26, 29, 35]

# Create boxplot
plt.boxplot(data, vert=False)
plt.title('Boxplot of Given Data')
plt.xlabel('Value')
plt.grid(True)
plt.show()

# Compute IQR to find outliers manually
q1 = np.percentile(data, 25)
q3 = np.percentile(data, 75)
iqr = q3 - q1
lower_bound = q1 - 1.5 * iqr
upper_bound = q3 + 1.5 * iqr

# Identify outliers
outliers = [x for x in data if x < lower_bound or x > upper_bound]

# Print results
print("Q1 (25th percentile):", q1)
print("Q3 (75th percentile):", q3)
print("IQR:", iqr)
print("Lower Bound:", lower_bound)
print("Upper Bound:", upper_bound)
print("Outliers:", outliers)
```

### **Question 8**: You are working as a data analyst in an e-commerce company. The marketing team wants to know if there is a relationship between advertising spend and daily sales.
### ● Explain how you would use covariance and correlation to explore this relationship.
### ● Write Python code to compute the correlation between the two lists:
### advertising_spend = [200, 250, 300, 400, 500]
### daily_sales = [2200, 2450, 2750, 3200, 4000]
### (Include your Python code and output in the code box below.)

## **Answer-**




### **Explanation:**

As a data analyst in an e-commerce company, to analyze the relationship between **advertising spend** and **daily sales**, we use:

### **(a). Covariance:**

* Covariance tells us the **direction** of the relationship.

  * **Positive covariance**: Both variables increase together.
  * **Negative covariance**: One increases while the other decreases.
* It **does not tell the strength** of the relationship.

### **(b). Correlation Coefficient (Pearson r):**

* Correlation shows both **direction and strength** of the relationship between two variables.
* It is a standardized value between **-1 and +1**:

  * **+1** = perfect positive correlation
  * **0** = no correlation
  * **–1** = perfect negative correlation
* More reliable than covariance for comparison.

**PYTHON CODE**


```python
# Python program to compute covariance and correlation
import numpy as np

# Given data
advertising_spend = [200, 250, 300, 400, 500]
daily_sales = [2200, 2450, 2750, 3200, 4000]

# Convert to NumPy arrays
x = np.array(advertising_spend)
y = np.array(daily_sales)

# Compute Covariance Matrix
cov_matrix = np.cov(x, y, bias=False)
covariance = cov_matrix[0][1]

# Compute Correlation Coefficient
correlation = np.corrcoef(x, y)[0][1]

# Print results
print("Covariance:", covariance)
print("Correlation Coefficient:", correlation)
```

**Conclusion:**
* The covariance is 87500, which is a positive value, meaning advertising spend and sales increase together.

* The correlation coefficient is approximately 0.9986, which is very close to +1, indicating a very strong positive linear relationship.

> This suggests that as advertising spend increases, daily sales also increase significantly — which can help the marketing team make data-driven investment decisions.

### Question 9: Your team has collected customer satisfaction survey data on a scale of 1-10 and wants to understand its distribution before launching a new product.
### ● Explain which summary statistics and visualizations (e.g. mean, standard deviation, histogram) you’d use.
### ● Write Python code to create a histogram using Matplotlib for the survey data:
### survey_scores = [7, 8, 5, 9, 6, 7, 8, 9, 10, 4, 7, 6, 9, 8, 7]
### (Include your Python code and output in the code box below.)

### **Answer:**

### **Explanation:**

To understand the **distribution** of customer satisfaction survey scores (rated on a 1–10 scale), the following **summary statistics and visualizations** are important:

### **Summary Statistics:**

1. **Mean** – average score (shows overall satisfaction level)
2. **Median** – middle score (helps detect skewed distribution)
3. **Mode** – most frequent score (shows common sentiment)
4. **Standard Deviation** – measures variability or spread of satisfaction levels
5. **Minimum and Maximum** – lowest and highest satisfaction score

### **Visualizations:**

1. **Histogram** – shows frequency distribution of satisfaction scores.
2. **Boxplot** (optional) – highlights median, quartiles, and outliers.
3. **Bar Chart** – to compare frequencies of each individual score (1–10).


### **Python Code-**

```python
# Python code to create a histogram of survey scores
import matplotlib.pyplot as plt

# Customer satisfaction survey data (scale 1–10)
survey_scores = [7, 8, 5, 9, 6, 7, 8, 9, 10, 4, 7, 6, 9, 8, 7]

# Create histogram
plt.hist(survey_scores, bins=range(4, 12), edgecolor='black', color='skyblue')
plt.title("Histogram of Customer Satisfaction Scores")
plt.xlabel("Survey Score")
plt.ylabel("Frequency")
plt.grid(True)
plt.show()
```
