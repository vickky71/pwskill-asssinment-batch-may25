1. What is the difference between a function and a method in Python?
ans--A function in Python is a block of reusable code that performs a specific task.

It’s defined using def (or lambda for anonymous functions).
It is independent and can be called directly.

Example:

def greet(name):
    return f"Hello, {name}"

print(greet("Vickky"))   # → Hello, Vickky

Method

A method is a function that belongs to an object (i.e., it’s defined inside a class).
It always takes at least one parameter: self (which represents the instance of the class).
It’s called using the dot notation on an object.

Example:

class Person:
    def __init__(self, name):
        self.name = name

    def greet(self):  # method
        return f"Hello, {self.name}"

p = Person("Vickky")
print(p.greet())  # → Hello, Vickky

6. Explain the concept of generators in Python and how they are defined

Ans-What is a Generator?

A generator is a function that produces a sequence of values lazily (one at a time, only when needed).

Instead of returning all values at once (like lists), it yields them using the yield keyword.

It automatically creates an iterator, so you can use next() or a for loop.

```python
# Using a Function with yield
def count_up_to(n):
    i = 1
    while i <= n:
        yield i      
        i += 1

gen = count_up_to(3)
print(next(gen)) 
print(next(gen)) 
print(next(gen)) 
```

```python
#Using Generator Expressions (like list comprehensions but with parentheses)
squares = (x*x for x in range(5)) 
print(next(squares)) 
print(next(squares)) 
print(next(squares)) 
```

7. What are the advantages of using generators over regular functions

   Ans --dvantages of Generators
1. Memory Efficiency

A regular function returning a list holds all values in memory at once.

A generator produces values one at a time (lazy evaluation), so it uses much less memory.

# Regular function
def make_list(n):
    return [i for i in range(n)]

# Generator
def make_gen(n):
    for i in range(n):
        yield i

print(sum(make_list(1000000)))   # uses a lot of memory
print(sum(make_gen(1000000)))    # memory-efficient

2. Lazy Evaluation

Values are generated on demand, not precomputed.

Great for working with large datasets or infinite streams.

def infinite_numbers():
    n = 1
    while True:
        yield n
        n += 1

gen = infinite_numbers()
print(next(gen))  # → 1
print(next(gen))  # → 2
print(next(gen))  # → 3

3. Performance (Speed in Iteration)

Since values are generated one by one, generators can be faster for iteration when you don’t need the whole result upfront.

4. Cleaner Code

Replaces complex iterator classes with simple generator functions.

# Instead of creating an Iterator class with __iter__ and __next__:
def count_up_to(n):
    i = 1
    while i <= n:
        yield i
        i += 1

5. Infinite / Streaming Data Support

A list cannot represent infinite values, but a generator can.

```python

def square(x):
    return x * x

# Lambda function
square_lambda = lambda x: x * x

print(square_lambda(5)) 
```

9. Explain the purpose and usage of the `map()` function in Python
   Ans --Purpose of map()

The map() function is used to apply a function to every item in an iterable (like list, tuple, etc.) and returns a map object (which is an iterator).
In short:It transforms each element of an iterable using a given function.
Syntax
map(function, iterable, ...)


function → the function to apply

iterable → one or more iterables (list, tuple, etc.)



```python
nums = [1, 2, 3, 4]
result = map(lambda x: x * 2, nums)
print(list(result))   # → [2, 4, 6, 8]
```

10. What is the difference between `map()`, `reduce()`, and `filter()` functions in Python

Ans -map()

Purpose: Apply a function to each element of an iterable.

Output: A new iterable (map object).

```python
nums = [1, 2, 3, 4]
result = map(lambda x: x*2, nums)
print(list(result))   # → [2, 4, 6, 8]
```

```python
nums = [1, 2, 3, 4, 5]
result = filter(lambda x: x % 2 == 0, nums)
print(list(result))   # → [2, 4]
```

reduce()

Purpose: Apply a function cumulatively to the elements of an iterable, reducing it to a single value.

Output: One final result.

Location: In functools module, so you need to import it.

```python
from functools import reduce

nums = [1, 2, 3, 4]
result = reduce(lambda x, y: x + y, nums)
print(result)   # → 10
```

1. Write a Python function that takes a list of numbers as input and returns the sum of all even numbers in 
the list

```python
def sum_of_evens(numbers):
    """
    Returns the sum of all even numbers in the input list.
    """
    return sum(num for num in numbers if num % 2 == 0)

nums = [1, 2, 3, 4, 5, 6]
print(sum_of_evens(nums))  # → 12 (2 + 4 + 6)
```

4. Write a Python function that checks if a given number is prime or not from 1 to 200.

```python
def is_prime(n):
    
    if n < 2 or n > 200:
        return False
    for i in range(2, int(n**0.5) + 1): 
        if n % i == 0:
            return False
    return True


for num in [1, 2, 15, 17, 200]:
    print(f"{num} is prime? {is_prime(num)}")
```

5. Create an iterator class in Python that generates the Fibonacci sequence up to a specified number of 
terms

```python
class Fibonacci:
    def __init__(self, n_terms):
        self.n_terms = n_terms   
        self.index = 0           
        self.a = 0               
        self.b = 1              

    def __iter__(self):
        return self

    def __next__(self):
        if self.index >= self.n_terms:
            raise StopIteration
        if self.index == 0:
            self.index += 1
            return self.a
        elif self.index == 1:
            self.index += 1
            return self.b
        else:
            self.a, self.b = self.b, self.a + self.b
            self.index += 1
            return self.b

fib_sequence = Fibonacci(10)
for num in fib_sequence:
    print(num, end=" ")  

```

6. Write a generator function in Python that yields the powers of 2 up to a given exponent

```python
def powers_of_two(max_exp):
    """
    Generator that yields powers of 2 from 0 to max_exp.
    """
    for i in range(max_exp + 1):
        yield 2 ** i


for value in powers_of_two(5):
    print(value, end=" ")  

```

def read_file_line_by_line(file_path):
    """
    Generator function that reads a file line by line.
    Yields each line as a string.
    """
    with open(file_path, 'r') as file:
        for line in file:
            yield line.rstrip('\n')  # Remove the newline character

# Example usage:
# Assuming 'example.txt' exists in the current directory
for line in read_file_line_by_line('example.txt'):
    print(line)


8. Use a lambda function in Python to sort a list of tuples based on the second element of each tuple

```python
# Sample list of tuples
data = [(1, 3), (4, 1), (2, 5), (3, 2)]

# Sort using lambda function (by second element)
sorted_data = sorted(data, key=lambda x: x[1])

print(sorted_data)  
# Output: [(4, 1), (3, 2), (1, 3), (2, 5)]
```

9. Write a Python program that uses `map()` to convert a list of temperatures from Celsius to Fahrenheit.

```python
# Function to convert Celsius to Fahrenheit
def celsius_to_fahrenheit(c):
    return (c * 9/5) + 32

celsius_temps = [0, 20, 37, 100]

fahrenheit_temps = list(map(celsius_to_fahrenheit, celsius_temps))

print(fahrenheit_temps)  
```

10. Create a Python program that uses `filter()` to remove all the vowels from a given string

```python
# Function to check if a character is not a vowel
def is_not_vowel(char):
    return char.lower() not in 'aeiou'

# Input string
text = "Hello, Python World!"

# Use filter() to remove vowels
result = ''.join(filter(is_not_vowel, text))

print(result)  
# Output: "Hll, Pythn Wrld!"
```
