## Question 1: What is hypothesis testing in statistics?

### Answer-**Hypothesis testing** is a statistical method used to make decisions or draw conclusions about a population based on sample data. It tests an assumption (called the **null hypothesis**, H₀) against an alternative hypothesis (H₁) to determine if there is enough evidence to support a specific claim.

### **Key Terms:**

* **Null Hypothesis (H₀):** The default assumption (e.g., no difference or effect).
* **Alternative Hypothesis (H₁):** The statement you want to test (e.g., there is a difference).
* **p-value:** Probability of obtaining the observed results if H₀ is true.
* **Significance Level (α):** Commonly set at 0.05. If p-value < α, reject H₀.


### **Steps:**

1. Formulate H₀ and H₁.
2. Choose a significance level (α).
3. Calculate the test statistic (e.g., t, z).
4. Find the p-value.
5. Make a decision:

   * If **p < α**, reject H₀.
   * If **p ≥ α**, fail to reject H₀.

### **Purpose:**

Hypothesis testing helps in making **data-driven decisions**, especially in research, business, and medical fields.

## Question 2: What is the null hypothesis, and how does it differ from the alternative hypothesis?

### Answer-The **null hypothesis (H₀)** is a statistical assumption that there is **no effect or difference** in the population. It serves as the starting point for hypothesis testing.

The **alternative hypothesis (H₁ or Ha)** is the opposite of the null. It states that there **is an effect or difference**, and is what the researcher aims to support.

###  **Example:**

* H₀: The average score = 70
* H₁: The average score ≠ 70


###  **Key Differences:**

| H₀ (Null Hypothesis)         | H₁ (Alternative Hypothesis) |
| ---------------------------- | --------------------------- |
| Assumes no effect/difference | Assumes effect/difference   |
| Tested directly              | Supported if H₀ is rejected |
| Conservative approach        | Research claim              |

### **Conclusion:**

Hypothesis testing evaluates if sample data provides strong enough evidence to **reject H₀** in favor of **H₁**.

### **Q3: Explain the Significance Level in Hypothesis Testing and Its Role in Deciding the Outcome**.

### Answer-The **significance level (α)** is a threshold used in hypothesis testing to decide whether to reject the null hypothesis (H₀). It represents the **maximum probability of making a Type I error**, i.e., rejecting H₀ when it is actually true.

Common values of α are **0.05, 0.01, or 0.10**, depending on how strict the test is.

### **Role in Decision-Making:**

* After calculating the **p-value** from sample data:

  * If **p-value ≤ α**, **reject H₀** (result is statistically significant).
  * If **p-value > α**, **fail to reject H₀** (insufficient evidence).

### **Example:**

If α = 0.05, you're accepting a 5% chance of wrongly rejecting a true null hypothesis. So, the test only rejects H₀ if the evidence is strong enough.

### **Conclusion:**

The significance level acts as a **cut-off point** that helps researchers control error risk and make **objective, data-based decisions** in hypothesis testing.

### **Q4: What are Type I and Type II Errors? Give Examples of Each?**

In hypothesis testing, errors can occur when making decisions about the null hypothesis (H₀). The two main types of errors are:

### **A. Type I Error (False Positive):**

* Occurs when we **reject a true null hypothesis**.
* It means we conclude there is an effect or difference when actually there is none.
* **Probability of Type I Error = α** (significance level).

**Example:**
A COVID-19 test shows a person is positive (infected), but in reality, they are not.

### **B. Type II Error (False Negative):**

* Occurs when we **fail to reject a false null hypothesis**.
* It means we miss a real effect or difference.
* **Probability of Type II Error = β**

**Example:**
A pregnancy test shows negative, but the woman is actually pregnant.

### **Comparison Table:**

| Error Type        | Reality (Truth) | Decision Made      | Outcome                           |
| ----------------- | --------------- | ------------------ | --------------------------------- |
| **Type I Error**  | H₀ is true      | H₀ is rejected     | False alarm (false positive)      |
| **Type II Error** | H₀ is false     | H₀ is not rejected | Missed detection (false negative) |


### **Conclusion:**

* Type I error leads to **false conclusions**, risking wrong actions.
* Type II error causes **missed discoveries**, risking inaction.
* Balancing both is key to making reliable decisions in hypothesis testing.

### **Q5: What is the Difference Between a Z-test and a T-test? When to Use Each?**

### Answer- A **Z-test** and a **T-test** are both statistical tests used to compare sample data with population data or between two samples, but they differ in usage based on **sample size** and **knowledge of population standard deviation**.

### **Z-Test:**

* Used when the **sample size is large** (n ≥ 30).
* **Population standard deviation (σ)** is known.
* Based on the **standard normal distribution**.
* More suitable for comparing sample mean to population mean in large datasets.

**Example:** Testing if average height of 100 students differs from population mean when σ is known.

### **T-Test:**

* Used when the **sample size is small** (n < 30).
* **Population standard deviation is unknown**.
* Based on the **t-distribution**, which has heavier tails.
* Types include **one-sample**, **independent two-sample**, and **paired t-tests**.

**Example:** Comparing average marks of two small student groups without knowing population σ.

### **Summary Table:**

| Feature                  | Z-Test                   | T-Test                        |
| ------------------------ | ------------------------ | ----------------------------- |
| Sample Size              | Large (n ≥ 30)           | Small (n < 30)                |
| Population Std. Dev. (σ) | Known                    | Unknown                       |
| Distribution Used        | Normal (Z)               | t-distribution                |
| Typical Use              | Proportions, large means | Small sample mean comparisons |



### **Conclusion:**

Use **Z-test** for large samples with known σ, and **T-test** for small samples or unknown σ. Choosing the right test ensures valid and reliable results.

## Question 6: Write a Python program to generate a binomial distribution with n=10 and p=0.5, then plot its histogram. (Include your Python code and output in the code box below.) Hint: Generate random number using random function.


```python
# Import necessary libraries
import numpy as np
import matplotlib.pyplot as plt

# Set parameters
n = 10       # number of trials
p = 0.5      # probability of success
size = 1000  # number of random samples

# Generate binomial distribution
binomial_data = np.random.binomial(n=n, p=p, size=size)

# Plot histogram
plt.hist(binomial_data, bins=range(n+2), edgecolor='black', align='left')
plt.title("Binomial Distribution Histogram (n=10, p=0.5)")
plt.xlabel("Number of Successes")
plt.ylabel("Frequency")
plt.grid(True)
plt.show()
```

## Question 7: Implement hypothesis testing using Z-statistics for a sample dataset in Python. Show the Python code and interpret the results. sample_data = [49.1, 50.2, 51.0, 48.7, 50.5, 49.8, 50.3, 50.7, 50.2, 49.6, 50.1, 49.9, 50.8, 50.4, 48.9, 50.6, 50.0, 49.7, 50.2, 49.5, 50.1, 50.3, 50.4, 50.5, 50.0, 50.7, 49.3, 49.8, 50.2, 50.9, 50.3, 50.4, 50.0, 49.7, 50.5, 49.9] (Include your Python code and output in the code box below.)


```python
import numpy as np
from scipy.stats import norm

# Sample data
sample_data = [49.1, 50.2, 51.0, 48.7, 50.5, 49.8, 50.3, 50.7, 50.2, 49.6,
               50.1, 49.9, 50.8, 50.4, 48.9, 50.6, 50.0, 49.7, 50.2, 49.5,
               50.1, 50.3, 50.4, 50.5, 50.0, 50.7, 49.3, 49.8, 50.2, 50.9,
               50.3, 50.4, 50.0, 49.7, 50.5, 49.9]

# Parameters
population_mean = 50       # μ₀ (assumed)
population_std = 0.5       # σ (assumed known)
n = len(sample_data)
sample_mean = np.mean(sample_data)

# Z-statistic
z = (sample_mean - population_mean) / (population_std / np.sqrt(n))

# p-value for two-tailed test
p_value = 2 * (1 - norm.cdf(abs(z)))

# Output
print(f"Sample Mean = {sample_mean:.4f}")
print(f"Z-statistic = {z:.4f}")
print(f"P-value = {p_value:.4f}")

# Decision
alpha = 0.05
if p_value < alpha:
    print("Reject the null hypothesis (H₀). Significant difference found.")
else:
    print("Fail to reject the null hypothesis (H₀). No significant difference.")
```

## Question 8: Write a Python script to simulate data from a normal distribution and calculate the 95% confidence interval for its mean. Plot the data using Matplotlib. (Include your Python code and output in the code box below.)


```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import t

# Step 1: Simulate data from a normal distribution
np.random.seed(0)  # for reproducibility
mean = 100         # population mean
std_dev = 15       # population standard deviation
n = 100            # sample size

data = np.random.normal(loc=mean, scale=std_dev, size=n)

# Step 2: Calculate sample mean and standard deviation
sample_mean = np.mean(data)
sample_std = np.std(data, ddof=1)  # sample standard deviation

# Step 3: Calculate 95% confidence interval
confidence = 0.95
alpha = 1 - confidence
df = n - 1  # degrees of freedom

# t-critical value
t_crit = t.ppf(1 - alpha/2, df)

# Margin of error
margin_error = t_crit * (sample_std / np.sqrt(n))

# Confidence interval
ci_lower = sample_mean - margin_error
ci_upper = sample_mean + margin_error

# Output
print(f"Sample Mean = {sample_mean:.2f}")
print(f"95% Confidence Interval = ({ci_lower:.2f}, {ci_upper:.2f})")

# Step 4: Plot histogram
plt.hist(data, bins=15, edgecolor='black', alpha=0.7)
plt.axvline(sample_mean, color='red', linestyle='--', label='Sample Mean')
plt.axvline(ci_lower, color='green', linestyle='--', label='95% CI Lower')
plt.axvline(ci_upper, color='green', linestyle='--', label='95% CI Upper')
plt.title("Normal Distribution Sample with 95% Confidence Interval")
plt.xlabel("Value")
plt.ylabel("Frequency")
plt.legend()
plt.grid(True)
plt.show()
```

## Question 9: Write a Python function to calculate the Z-scores from a dataset and visualize the standardized data using a histogram. Explain what the Z-scores represent in terms of standard deviations from the mean. (Include your Python code and output in the code box below.)


```python
import numpy as np
import matplotlib.pyplot as plt

# Function to calculate Z-scores
def calculate_z_scores(data):
    mean = np.mean(data)
    std = np.std(data)
    z_scores = (data - mean) / std
    return z_scores

# Sample dataset (random normal data)
np.random.seed(42)
data = np.random.normal(loc=100, scale=15, size=100)

# Calculate Z-scores
z_scores = calculate_z_scores(data)

# Print first 5 Z-scores
print("First 5 Z-scores:")
print(z_scores[:5])

# Plot histogram of Z-scores
plt.hist(z_scores, bins=15, edgecolor='black', alpha=0.7)
plt.title("Histogram of Z-scores (Standardized Data)")
plt.xlabel("Z-score")
plt.ylabel("Frequency")
plt.grid(True)
plt.show()
```
