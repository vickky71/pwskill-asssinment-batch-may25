1. What is the difference between a function and a method in Python with example ?

  Function: A function is defined using the def keyword and exists independently. It is not tied to any specific object.
  Method: A method is a function that is associated with an object and is called on that object. It is defined within a class.

```python
#Function:
def greet(name):
    return "Hello " + name

print(greet("Vickky"))  
```

```python
 # Method:
class Person:
    def __init__(self, name):
        self.name = name

    def greet(self):  # This is a method
        return "Hello " + self.name

p = Person("Vickky")
print(p.greet()) 
```

5. What are iterators in Python and how do they differ from iterables?

An iterable is any Python object capable of returning its elements one at a time. It can be passed to the iter() function to create an iterator.
Examples: Lists, tuples, strings, dictionaries, sets, and custom objects implementing the __iter__() method.
Key Feature: Iterables do not maintain any state about their iteration. They can be iterated over multiple times (e.g., using a for loop).

An iterator is an object that represents a stream of data. It is created by calling iter() on an iterable and is used to fetch elements one at a time using the next() function.Iterators maintain an internal state to keep track of where they are in the iteration. Once exhausted, they cannot be reused unless explicitly reset.

```python
#iterable
my_list = [1, 2, 3]  # A list is an iterable
for item in my_list:
    print(item)
```

```python
#iterator
my_list = [1, 2, 3]
iterator = iter(my_list)  

print(next(iterator)) 
print(next(iterator))  
print(next(iterator))  

```

6.  Explain the concept of generators in Python and how they are defined.

A generator function is a special type of function that returns an iterator object. Instead of using return to send back a single value, generator functions use yield to produce a series of results over time. This allows the function to generate values and pause its execution after each yield, maintaining its state between iterations.

```python
def even_numbers(limit):
    for i in range(limit):
        if i % 2 == 0:
            yield i

for num in even_numbers(10):
    print(num)
```

7. What are the advantages of using generators over regular functions?

The use of generators in Python offers various benefits that we have explained below:

A normal function creates an entire sequence in memory to return a sequence from the list. If the number of items in the sequence is large, it can consume ample memory. However, a generator function is memory-friendly as it calculates the value and produces one item at a time.

10. What is the difference between `map()`, `reduce()`, and `filter()` functions in Python?

The map() function applies a given function to each item of an iterable (like a list) and returns a map object (an iterator). The syntax is:   map(function, iterable)

The filter() function constructs an iterator from elements of an iterable for which a function returns true. The syntax is:filter(function, iterable)

The reduce() function applies a function of two arguments cumulatively to the items of an iterable, from left to right, to reduce the iterable to a single value. It is part of the functools module. The syntax is:

reduce(function, iterable[, initial])

```python
#The map() function

nums = [1, 2, 3, 4]

squares = map(lambda x: x * x, nums)

print(list(squares)) 
```

```python
#The filter() function

nums = [1, 2, 3, 4, 5, 6]
evens = list(filter(lambda x: x % 2 == 0, nums))
print(evens)  
```

```python
#The reduce() function

from functools import reduce

nums = [1, 2, 3, 4]
product = reduce(lambda x, y: x * y, nums)
print(product) 
```

4. Write a Python function that checks if a given number is prime or not from 1 to 200.

```python
def is_prime(n):
    if n < 2:
        return False  
    for i in range(2, int(n**0.5) + 1): 
        if n % i == 0:
            return False
    return True

for num in range(1, 201):
    if is_prime(num):
        print(num, "is a prime number")

```

5. Create an iterator class in Python that generates the Fibonacci sequence up to a specified number of 
terms

```python
class FibonacciIterator:
    def __init__(self, max_terms):
        self.max_terms = max_terms
        self.count = 0
        self.a = 0
        self.b = 1

    def __iter__(self):
        return self

    def __next__(self):
        if self.count >= self.max_terms:
            raise StopIteration
        if self.count == 0:
            self.count += 1
            return self.a
        elif self.count == 1:
            self.count += 1
            return self.b
        else:
            self.a, self.b = self.b, self.a + self.b
            self.count += 1
            return self.b


fib = FibonacciIterator(10)
for num in fib:
    print(num)

```

6. Write a generator function in Python that yields the powers of 2 up to a given exponent.

```python
def powers_of_two(max_exponent):
    for exp in range(max_exponent + 1):
        yield 2 ** exp

for num in powers_of_two(5):
    print(num)
```

7. Implement a generator function that reads a file line by line and yields each line as a string.

```python
def read_file_line_by_line(file_path):
    """
    A generator function to read a file line by line.

    Args:
        file_path (str): Path to the file to be read.

    Yields:
        str: Each line of the file as a string.
    """
    try:
        with open(file_path, 'r', encoding='utf-8') as file:
            for line in file:
                yield line.rstrip('\n')  # Remove trailing newline characters
    except FileNotFoundError:
        print(f"Error: The file at '{file_path}' was not found.")
    except Exception as e:
        print(f"An error occurred: {e}")
```

9. Write a Python program that uses `map()` to convert a list of temperatures from Celsius to Fahrenheit.

```python
# Celsius to Fahrenheit conversion formula: F = (C * 9/5) + 32

def celsius_to_fahrenheit(celsius):
    return (celsius * 9/5) + 32

celsius_list = [0, 20, 30, 37, 100]

fahrenheit_list = list(map(celsius_to_fahrenheit, celsius_list))

print("Fahrenheit temperatures:", fahrenheit_list)
```

10. Create a Python program that uses `filter()` to remove all the vowels from a given string

```python
def remove_vowels(input_str):
    vowels = 'aeiouAEIOU'
    return ''.join(filter(lambda char: char not in vowels, input_str))

text = "Hello World, Python is amazing!"
result = remove_vowels(text)
print("String without vowels:", result)
```

11. Write a Python program, which returns a list with 2-tuples. Each tuple consists of the order number and the 
product of the price per item and the quantity. The product should be increased by 10,- â‚¬ if the value of the 
order is smaller than 100,00 â‚¬

```python
def process_orders(orders):
    result = []
    for order in orders:
        order_no, price_per_item, quantity = order
        total = price_per_item * quantity
        if total < 100:
            total += 10  # Add â‚¬10 if the order value is less than â‚¬100
        result.append((order_no, total))
    return result

# Example list of orders: (order_no, price_per_item, quantity)
orders = [
    (1, 20.0, 2),   # 40 â†’ 40 + 10 = 50
    (2, 50.0, 3),   # 150 â†’ no fee
    (3, 15.0, 5),   # 75 â†’ 75 + 10 = 85
    (4, 100.0, 1),  # 100 â†’ no fee
]

# Process and print the results
processed = process_orders(orders)
print("Processed orders:", processed)

```
